# Change DataFrame Background Color
Use df.style to change the DataFrame background to a green gradient based on cell value.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(0, 100, size=(15, 9)), columns=list("ABCDEFGHI"))
df.style.background_gradient(cmap="Greens")
```
