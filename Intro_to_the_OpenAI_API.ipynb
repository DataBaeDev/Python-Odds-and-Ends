## Create an OpenAI instance
Make sure to create an account at https://platform.openai.com/docs/overview and obtain an API key

```python
client = OpenAI(api_key= openai_key)
```

## Define a chat function that uses the *gpt-3.5-turbo-1106* model

```python
"""
Send a message to the OpenAI gpt-3.5-turbo-1106 model and return its response.
Parameters: message (str): A string containing the user's message to the AI.
Returns:  str: The text response generated by the GPT-3.5 model.
"""

def chat(user_msg):
  response = client.chat.completions.create(
  model="gpt-3.5-turbo-1106",
  messages=[
      {"role": "system", "content": "You are travel agent that concisely answers questions about business and leisure destinations from travelers."},
      {"role": "user", "content": f"{user_msg}"}
  ]
)
  gpt_response_text = response.choices[0].message.content
  return gpt_response_text
```

## Prompt the user to ask a question and return a response using the chat function

```python
chat(input("Hello, what are your travel needs today?"))
```
