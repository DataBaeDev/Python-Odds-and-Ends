# Python File Mover

This script moves all files in a directory to a folder called "completed" and places a timestamp on each file.

```python
import os
import shutil  ## used for moving files
import datetime as dt
```

```python
# initialize directories
source_dir = r"C:\Users\ronis\Desktop\Sample Directory"

# Create 'completed' directory if it doesn't exist
completed_dir = os.path.join(source_dir, "completed")
if not os.path.exists(completed_dir):
    os.makedirs(completed_dir)

# Create datetime stamp
timestamp = dt.datetime.now().strftime("%Y%m%d_%H%M")

# determine file type to check for in the directory
file_ext = ".txt"
```

```python
# Loop through files in the source directory
# Check for a specific file type
# Split filename and extension, create the new filename, then move file

for file in os.listdir(source_dir):

    file_path = os.path.join(source_dir, file)

    if os.path.isfile(file_path):
        name, ext = os.path.splitext(file)

        if ext == file_ext:
            new_filename = f"{name}_{timestamp}{ext}"
            new_file_path = os.path.join(completed_dir, new_filename)

            shutil.move(file_path, new_file_path)

            print(f"Moved: {file} -> completed/{new_filename}")

print("All files have been moved and date stamped.")
```
