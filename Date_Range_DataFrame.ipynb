# Create a Date Range DataFrame
This DataFrame can serve a similar purpose as a SQL Date Dimension Table

```python
import pandas as pd
```

```python
# Create a transactional DataFrame
transactions = {
    'Transaction ID': ['T001', 'T002', 'T003', 'T004', 'T005'],
    'Start Date': ['2023-06-01', '2023-06-05', '2023-06-10', '2023-06-15', '2023-06-20'],
    'End Date': ['2023-06-03', '2023-06-08', '2023-06-13', '2023-06-18', '2023-06-25']
}

transactions_df = pd.DataFrame(transactions)
print(transactions_df)
```

Extract the earliest and latest transaction dates to generate a date range for use in the date DateFrame

```python
earliest_start_date = pd.to_datetime(transactions_df['Start Date']).min()
latest_end_date = pd.to_datetime(transactions_df['End Date']).max()

print(f"Date Range: {earliest_start_date} - {latest_end_date}")
```

Generate date DataFrame

```python
# Create a date range between the earliest start date and latest end date
date_range = pd.date_range(start=earliest_start_date, end=latest_end_date, freq='D')

# Create the second dataframe with the date column
date_df = pd.DataFrame({'Date': date_range})

print(date_df.head())
```

Create a final DataFrame that counts the number of transactions in progress on each date.

```python
transaction_summary_df = date_df.copy()

# Add a column that counts the number of transactions in progress for each date
transaction_summary_df['Transactions in Progress'] = transaction_summary_df['Date'].apply(lambda date: ((pd.to_datetime(transactions_df['Start Date']) <= date) & (pd.to_datetime(transactions_df['End Date']) >= date)).sum())

print(transaction_summary_df)
```
